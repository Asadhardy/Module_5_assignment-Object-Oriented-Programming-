```python
'''
Q6. Describe the five types of inheritance in Python. Provide a simple example of multiple inheritance.
Answer:
    In python, there are five types of inheritance , which allows  a class to inherit properties,
    and methods from another class.
    Each typer serve a different purpose based on the relationship between classes .
    The five types are as following:
1.Single Inheritance:
    A class inherits from one base class,
Example
'''
class Parent:
    def show(self):
        print("this is a parent class.")
class Child(Parent):
    pass
child = Child()
child.show()

'''
2.Multiple Inheritance:
    A class inherits from more than one base class.
Example:
'''
class Parent1():
    def show1(self):
        print("This is the parent 1 class")
class Parent2:
    def show2(self):
        print("THis is the parent 2 class")
#Inheriting from 2 parent classs
class Child(Parent1, Parent2):
    pass
child1 = Child()
child1.show1() # this will show method of parent 1
child1.show2() # and this will show the method of parent 2
'''
3.Multilevel Inheritance
    A class inherits from a derived class creating a chain of inheritance.
Exmaple: 
'''
class Grandparent:
    def show(self):
        print("This is the grandparent's class")
class Parent(Grandparent):
    pass
class Child(Parent):
    pass
child = Child()
child.show()
'''
4.Hierarchical Inheritance:
    Multiple classes inherit from the same base class.
Example:
'''
class Parent:
    def show(self):
        print("This is the parent's hierarchical class")
class Child1(Parent):
    pass
class Child2(Parent):
    pass
child1 = Child1()
child1.show()
child2 = Child2()
child2.show()
# both will have the same method (show), incuding their own saperated.
'''
5.Hybrid Inheritance
    A combination of multiple inheritance types (like multilevel and multiple inheritance).
Example:
'''
# Base class
class Grandparent:
    def show(self):
        print("This is the grandparent class.")

# Inheriting from Grandparent (Multilevel Inheritance)
class Parent1(Grandparent):
    def show1(self):
        print("This is Parent1, inheriting from Grandparent.")

# Another parent classs
class Parent2:
    def show2(self):
        print("This is Parent2, a separate parent class.")

# Child class inherits from both Parent1 and Parent2 (Multiple Inheritance)
class Child(Parent1, Parent2):
    def show3(self):
        print("This is the child class, inheriting from both Parent1 and Parent2.")

# Creatingg an object of Child class
child = Child()
child.show()   # Inherited from Grandparent (Multilevel)
child.show1()  # Inherited from Parent1 (Multilevel)
child.show2()  # Inherited from Parent2 (Multiple)
child.show3()  # Child class method

'''
All five different types of inheritance examples are above, including the multiple which was asked for at point 2

Insights:
    Inheritance in python provides a way for a class to derived attributes and method from another class, 
    enabling code reusability and logical hierarchy, whether it is a single or multiple inheritance,
    each type offers flexibility in structuring programs, with multiple inheritance allowing a class to
    inherit from more than one parent class.
    However, in complex cases, careful management of the method resolution order(MRO) is necessary to avoid conflicts,
    means, which will be the first class is inherited(positionally) will be executed.

'''
print()
```

```python
'''
Q9.Demonstrate polymorphism by creating a function that can work with different shape objects to calculate
and print their areas.?
Answer Code:
'''
def square(size = 8):
    area = size ** 2
    print(f"The Area of a square is: {area}")
def rectangle(length = 15,breadth = 30):
    area = length*breadth
    print(f"The Area of a rectangle is: {area}")
def circle(radius = 7):
    area = 3.141 * radius ** 2
    print(f"The Areas if the circle is: {area}")

def calculate_area(shape, **kwargs):
    shape = shape.lower()
    if shape == "circle":
        circle(**kwargs)
    elif shape == "rectangle":
        rectangle(**kwargs)
    elif shape == "square":
        square(**kwargs)
    else:
        return f"Please enter an valid shape like Circle, Rectangle or Square"



calculate_area("Circle") #default radius
calculate_area("Circle", radius=9) #radius given in the agrument, and it will override the default radius
calculate_area("rectangle") # with default radius 
calculate_area("rectangle",length=14, breadth=28) #length=14, breadth=28) without detault length and breadth
calculate_area("square") #with default size
calculate_area("square", size=9) #with size in the agruement

'''
Insights:
    The function calculate_area takes a shape as input and calls the appropriate area calculation function based on the input.
    You can pass additional arguments like radius, length, or size using **kwargs to override the default values.

    If you pass "Circle" as the shape, it will calculate the area of a circle (default or with a given radius).
    Similarly, for "Rectangle" or "Square", the corresponding functions will calculate their areas using default or provided dimensions.
    This setup is flexible and allows you to calculate the area for different shapes with ease.
'''
```

```python
'''
Q13. Explain the concept of the Diamond Problem in multiple inheritance. How does Python resolve it?
Answer:
    The diamond problem occurs in multiple inheritance when a class inherits from two or more classes that ,
    share a common base class.
    This created uncertainty because it is unclear which path should be followed to inherit methods or attributes
    from the shared base class.
    The structure forms a diamond shape in the inheritance hierarchy .
    Hence, It is know as "diamond problem"
Example:
'''
class A:
    def show(self):
        print("A class")
class B(A):# inheriting A class
    def show(self):
        print("B class")
class C(A):
    def show(self):
        print("C class")
class D (C, B): #Which one is positionally first , that will be evaluated
    # D class will inherits from both B and C class, which is inherit from A
    pass

d = D()
d.show() # output of the D class is 'C class'
'''
1.In the above example , class D inherits from both B and C, and both of these classes inherit from A.
2.The question arises when you class d.show, should python use the method from B or C ?
#This uncertainty is call diamond problem.

How to resolve it:
    python resolves the diamond problem using method Resolution Order(MRO).
    Python follows the C3 Linearization Algorithm to determine the order in which classes are searched for the,
    methods or attribute.
    This ensures that the search path is consistent, follows a clear hierarchy and avoids uncertainty.

In simple language:
    Python first checks the method in the current class (D), it method is not there,
    Then it moves to the first parent (B), and if not found , it will move to the second parent (C) in search of method 
    and finally to the base class (A)

You can check MRO using the mro() method or the __mro__ attribute.
Example to check the MRO:
'''
print(D.mro())
'''
Explanation:
    The Mro determined that the search order for method show() is D -> B -> C -> A.
    So, d.show() will call the show() method from the class B first , 
    as it appears before C in the MRO.(positionally in while inheriting the classes)
    
Insights:
    The diamond problem in multiple inheritance can create uncertainty when multiple classes inherit from a shared,
    ancestor.
    Python resolves this issue using the MRO, ensuring a clear and consistent method lookup order, 
    This prevents confusion and maintains the intergrity of the inheritance hierarchy
'''
print()
```
